```python
import pandas as pd
import numpy as np
```

```python
list1 = []
for i in range(2,10):
    locals()[f'data{i}'] = pd.read_excel('/Users/malek/Documents/LSEG Data/Korean Returns.xlsx',sheet_name=f'Sheet{i}',index_col=0)
    list1.append(locals()[f'data{i}'])
```

```python
# list1 contains 8 DataFrames with identical columns but with different time periods
list1 = [i.apply(pd.to_numeric,errors='coerce') for i in list1]
list1 = [i.dropna(axis=1,how='all') for i in list1]
list1 = [i.set_index(pd.to_datetime(i.index,yearfirst=True)) for i in list1]
```

```python
Korean_Returns = pd.concat(list1, axis=0).sort_index()
Korean_Returns = Korean_Returns.pct_change()
Korean_Returns = Korean_Returns.replace({0:np.nan})
Korean_Returns.columns = Korean_Returns.columns.str.replace(' - TOT RETURN IND','')
Korean_Returns
```
